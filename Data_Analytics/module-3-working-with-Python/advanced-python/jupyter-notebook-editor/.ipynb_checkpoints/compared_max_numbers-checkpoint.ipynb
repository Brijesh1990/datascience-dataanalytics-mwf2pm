```python
# create a banking app details 
# or check max numbers among three numbers 
n1=int(input('Enter N1 Numbers :'))
n2=int(input('Enter N2 Numbers :'))
n3=int(input('Enter N3 Numbers :'))
if n1>n2 and n1>n3:
    print("N1 is greter Numbers")
elif n2 > n1 and n2>n3:
    print("N2 is greter Numbers")
elif n3 > n2 and n3>n1:
    print("N3 is greter Numbers")
else:
    print("You enter a wrong number to compared values ")




```
